## **PRODUCTS DATA**

```python
path = "/content/drive/MyDrive/Jobs/Kellogg's/product.csv"
data = pd.read_csv(path)
print('Date Shape: {}'.format(data.shape))

data.tail()
```

```python
# data.info()
print('Total Null Values: {}\n'.format(data.isnull().sum().sum()))

data.isnull().sum()
```

```python
print('Total No.of Products: {}'.format(len(data.product_id.unique())))
print('Total No.of Brands  : {}'.format(len(data.brand_id.unique())))
```

```python
print('Presence of Competitive Brands\n{}'.format(data.K_flag.value_counts()))
```

```python
fig = go.Figure()
fig.add_trace(
    go.Heatmap(
        x = data.corr().columns,
        y = data.corr().index,
        z = np.array(data.corr()),
        text=data.corr().values,
        texttemplate='%{text:.2f}'
    )
)
fig.update_layout(height=500, width=700, xaxis_title="Feature Variables", yaxis_title="Feature Variables", title='Correlation Plot')
fig.show()
```

```python
brands = pd.DataFrame(data.K_flag.value_counts()).reset_index()
brands.head()
```

```python
fig = px.bar(brands, x = brands.index, y = brands.K_flag, color = brands.index,  barmode='group')
fig.update_xaxes(rangeslider_visible=False, showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_layout(height=600, width=500, xaxis_title="Presence of Competitive Brands", yaxis_title="Frequency", title_text="Frequency of Brands w.r.t Products") 
fig.show()
```

```python
brands = pd.DataFrame(data.groupby(['brand_id'])['K_flag'].first()).reset_index()
brands.tail()
```

```python
fig = px.bar(brands, x = brands.brand_id, y = brands.K_flag, color = brands.brand_id,  barmode='group')
fig.update_xaxes(rangeslider_visible=False, showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_layout(height=500, width=1000, xaxis_title="Frequency of Competitive Brands", yaxis_title="Presence of Competitive Brands", title_text="Presence of Competitive Brans vs Brand ID") 
fig.show()
```

## **CUSTOMER SUPPLEMENT**

```python
path = "/content/drive/MyDrive/Jobs/Kellogg's/customer_supplement.csv"
data = pd.read_csv(path)
print('Date Shape: {}'.format(data.shape))

data.head(10)
```

```python
# data.info()
print('Date Shape: {}'.format(data.shape))
print('Total Null Values (Before): {}'.format(data.isnull().sum().sum()))

data.dropna(inplace = True)
data.income = data.income_from.apply(lambda x: int(x))
data.drop(['income_from', 'income_to'], axis = 1, inplace = True)

print('\nTotal Null Values (After): {}'.format(data.isnull().sum().sum()))
print('Date Shape: {}\n'.format(data.shape))

data.head(10)
```

```python
Qualification = pd.DataFrame(data.groupby(['qualification'])['income'].mean()).reset_index()

fig = px.bar(Qualification, x = Qualification.qualification, y = Qualification.income, color = Qualification.income,  barmode='group')
fig.update_xaxes(rangeslider_visible=False, showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_layout(height=600, width=800, xaxis_title="Qualifications", yaxis_title="Income", title_text="Income w.r.t Qualifications of Customers") 
fig.show()
```

```python
Age = pd.DataFrame(data.groupby(['age'])['income'].mean()).reset_index()
Age = Age[Age.age < 100]

fig = px.bar(Age, x = Age.age, y = Age.income, color = Age.income)
fig.update_xaxes(rangeslider_visible=False, showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_layout(height=600, width=1400, xaxis_title='Age', yaxis_title="Income", title_text="Income vs Age of Customers") 
fig.show()
```
